```python
import pandas as pd

dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv',sep=';')
dados.head(5)
```

```python
#Criando uma coluna numérica
dados['Valor_por_Mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_Ano'] = dados['Valor']*12 + dados['IPTU']
dados.head(5)
```

Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + " em " + dados['Bairro'] + " com " + \
                    dados['Quartos'].astype(str) + ' quarto(s) ' + \
                    ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.' 
# astypes() muda o dado para outro tipo que vc queira, mudei para str
```

```python
dados.head(2)
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")

#dados.query('Suite > 1').index
```

```python
dados.head(3)
```

```python
dados.to_csv('data/dados_completos_dev.csv',sep=';',index=False)
```

```python
url2 = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'

dados = pd.read_csv(url2,sep=',')
```

```python
dados.fillna(0)
```

```python
dados['Pontos_Extras'] = dados['Notas'].apply(lambda x: x*0.4)
```

```python
dados.head(5)
```

```python
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_Extras']
```

```python
dados.head(4)
```

```python
dados['Aprovado_final'] = dados['Notas_finais'].apply(lambda x: True if x >= 6 else False)
```
